```python
# Use your Python skills to create a visualization here.

# Scatter Plot
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style

df = pd.read_csv("books.csv", on_bad_lines='skip')

df.columns = df.columns.str.strip()

plt.figure(figsize=(10,6))
plt.scatter(df['num_pages'], df['average_rating'], alpha=0.5, color='purple')
plt.title("Book Length vs Average Rating")
plt.xlabel("Number of Pages")
plt.ylabel("Average Rating")
plt.show()
```

```python
# Use your Python skills to create a visualization here.

top_authors = df['authors'].value_counts().head(5)

plt.figure(figsize=(8,8))
plt.pie(top_authors, labels=top_authors.index, autopct='%1.1f%%', startangle=140)
plt.title("Top 5 Authors by Number of Books")
plt.show()
```
